```python
#Consommation par pays

df_fuel_conso = pd.read_csv("fossil-fuel-consumption-per-capita.csv")
df_fuel_conso = df_fuel_conso.rename(columns={"Oil (megawatt-hours)": "oil_mwh","Coal (megawatt-hours)": "coal_mwh" ,"Natural gas (megawatt-hours)": "gas_mwh" })
```

```python
df_fuel_conso_India = df_fuel_conso[df_fuel_conso['Entity'] == "India"]
df_fuel_conso_China = df_fuel_conso[df_fuel_conso['Entity'] == "China"]
```

```python
sns.set()
x=df_fuel_conso_India.Year  # pour recuperer les annees de 1910  2019
y=[df_fuel_conso_India.coal_mwh, df_fuel_conso_India.oil_mwh, df_fuel_conso_India.gas_mwh ]

plt.figure(figsize=(12, 8))
plt.stackplot(x,y, labels=['Coal','Oil','Gas'])
plt.legend(loc='upper left')
plt.title ('Per capita fossil fuel consumption in India')
plt.ylabel("Mwh ")
plt.xlabel("Year")

plt.show()
```
