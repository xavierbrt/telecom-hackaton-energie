```python
import pandas as pd
```

```python
xl = pd.ExcelFile("bp-stats-review-2019-all-data.xlsx")
xl.sheet_names  # see all sheet names
```

```python
def get_data_consumption(sheet_name):
    df = pd.read_excel("bp-stats-review-2019-all-data.xlsx", sheet_name=sheet_name, header=2,nrows=108, usecols=range(55))
    df = df.dropna(axis=0, how='all', thresh=None, subset=None, inplace=False)
    df = df[~df['Million tonnes oil equivalent'].str.contains("Total")]
    df = df.set_index("Million tonnes oil equivalent")
    df.index.name = None
    return df
```

```python
df_oil_cons = get_data_consumption("Oil Consumption - Mtoe")
df_gas_cons = get_data_consumption("Gas Consumption - Mtoe")
df_coal_cons = get_data_consumption("Coal Consumption - Mtoe")
df_nucl_cons = get_data_consumption("Nuclear Consumption - Mtoe")
df_hydr_cons = get_data_consumption("Hydro Consumption - Mtoe")
df_renew_cons = get_data_consumption("Renewables - Mtoe")
```
