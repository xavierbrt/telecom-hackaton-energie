```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "world-population-by-world-regions-post-1820.csv"

df_population = pd.read_csv(url)
```

```python
df_popu_growth= df_population[df_population['Code'].isna()]
df_popu_growth_from_1900 = df_popu_growth[df_popu_growth['Year'] > 1900].sort_values(by = "Population")
df_popu_growth_from_1900= df_popu_growth_from_1900[df_popu_growth_from_1900['Entity']!= "Saint Barthlemy"] # 
```

```python
# Data
pop_Asia = df_popu_growth_from_1900[df_popu_growth_from_1900['Entity'] == 'Asia'].Population
pop_Africa = df_popu_growth_from_1900[df_popu_growth_from_1900['Entity'] == 'Africa'].Population
pop_Europe = df_popu_growth_from_1900[df_popu_growth_from_1900['Entity'] == 'Europe'].Population
pop_Latin_America = df_popu_growth_from_1900[df_popu_growth_from_1900['Entity'] == 'Latin America'].Population
pop_North_America = df_popu_growth_from_1900[df_popu_growth_from_1900['Entity'] == 'North America'].Population
pop_Oceania = df_popu_growth_from_1900[df_popu_growth_from_1900['Entity'] == 'Oceania'].Population

# Plot
sns.set()
x=df_popu_growth_from_1900[df_popu_growth_from_1900['Entity'] == 'Asia'].Year  # pour recuperer les annees de 1910  2019
y=[pop_Asia, pop_Africa, pop_Europe,pop_North_America, pop_Latin_America, pop_Oceania ]

plt.figure(figsize=(12, 8))
plt.stackplot(x,y, labels=['Asia','Africa','Europe','North America', 'Latin America', 'Oceania'])
plt.legend(loc='upper left')
plt.title ('Population growth by world region')
plt.ylabel("Population ")
plt.xlabel("Year")

plt.show()
```
