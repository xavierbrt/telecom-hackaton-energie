```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#CSV
df_ren_investment = pd.read_csv("renewable-energy-investment.csv")
df_ren_investment = df_ren_investment.rename(columns={"Investment in Renewables by Region (IRENA (2016)) (international-$)": "investment"})

#DATA
year = df_ren_investment[df_ren_investment['Entity'] == "India"].Year
df_ren_investment_Europe = df_ren_investment[df_ren_investment['Entity'] == "Europe"].investment
df_ren_investment_USA = df_ren_investment[df_ren_investment['Entity'] == "United States of America"].investment
df_ren_investment_India = df_ren_investment[df_ren_investment['Entity'] == "India"].investment
df_ren_investment_China = df_ren_investment[df_ren_investment['Entity'] == "China"].investment

sns.set()
x=year  # pour recuperer les annees de 1910 à 2019
y=[df_ren_investment_USA,df_ren_investment_India, df_ren_investment_China, df_ren_investment_Europe ]

#PLOT
plt.figure(figsize=(12, 8))
plt.stackplot(x,y, labels=['USA','India','China','Europe'])
plt.legend(loc='upper left')
plt.title ('Renewable Energy Investment')
plt.ylabel("$250 billion ")
plt.xlabel("Year")
plt.show()
```
